```python
#splitting the train/validation/test split  of 60/20/20 distribution
from sklearn.model_selection import train_test_split

# First split: 60% training and 40% remaining (for validation and test)
train, temp = train_test_split(df, train_size=0.6, random_state=1)

# Second split: 20% validation and 20% test from the remaining 40%
val, test = train_test_split(temp, test_size=0.5, random_state=1)

# Check the shapes to confirm the split
print("Train set:", train.shape)
print("Validation set:", val.shape)
print("Test set:", test.shape)

#Train set: (3000, 16)
#Validation set: (1000, 16)
#Test set: (1000, 16)
```

Q1: Let's train a decision tree regressor to predict the jamb_score variable.

Train a model with max_depth=1.
Which feature is used for splitting the data?

study_hours_per_week
attendance_rate
teacher_quality
distance_to_school


```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeRegressor

# Remove target variable from DataFrames
X_train_df = train.drop(columns=['jamb_score'])
X_val_df = val.drop(columns=['jamb_score'])
X_test_df = test.drop(columns=['jamb_score'])

# Convert each DataFrame to a list of dictionaries
train_dict = X_train_df.to_dict(orient='records')
val_dict = X_val_df.to_dict(orient='records')
test_dict = X_test_df.to_dict(orient='records')

# Initialize DictVectorizer
vec = DictVectorizer(sparse=True)

# Fit and transform the train set, then transform the validation and test sets
X_train = vec.fit_transform(train_dict)
X_val = vec.transform(val_dict)
X_test = vec.transform(test_dict)

# Extract y_train and y_val separately
y_train = train['jamb_score']
y_val = val['jamb_score']

# Train the Decision Tree with max_depth=1
model = DecisionTreeRegressor(max_depth=1, random_state=1)
model.fit(X_train, y_train)

# Retrieve the feature used for the split
feature_index = model.tree_.feature[0]
feature_name = vec.feature_names_[feature_index]

print("Feature used for splitting:", feature_name)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
rf_model.fit(X_train, y_train)

# Predict on the validation set
y_pred = rf_model.predict(X_val)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_val, y_pred))

print("RMSE on the validation data:", rmse)
```

Q3. Now let's experiment with the n_estimators parameter

Try different values of this parameter from 10 to 200 with step 10.
Set random_state to 1.
Evaluate the model on the validation dataset.
After which value of n_estimators does RMSE stop improving? Consider 3 decimal places for calculating the answer.


```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize variables
n_estimators_values = range(10, 201, 10)  # Values from 10 to 200 with step 10
rmse_values = []  # To store the RMSE for each n_estimators

# Loop through each value of n_estimators
for n_estimators in n_estimators_values:
    rf_model = RandomForestRegressor(n_estimators=n_estimators, random_state=1, n_jobs=-1)
    rf_model.fit(X_train, y_train)

    # Predict on the validation set
    y_pred = rf_model.predict(X_val)

    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    rmse_values.append(rmse)

    print(f"n_estimators: {n_estimators}, RMSE: {rmse}")

# Find the point where RMSE stops improving
for i in range(1, len(rmse_values)):
    if round(rmse_values[i], 3) == round(rmse_values[i-1], 3):
        print(f"RMSE stops improving after n_estimators = {n_estimators_values[i-1]}")
        break
```

Question 4
Let's select the best max_depth:

Try different values of max_depth: [10, 15, 20, 25]
For each of these values,
try different values of n_estimators from 10 till 200 (with step 10)
calculate the mean RMSE
Fix the random seed: random_state=1
What's the best max_depth, using the mean RMSE?

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize variables
max_depth_values = [10, 15, 20, 25]
n_estimators_values = range(10, 201, 10)  # Values from 10 to 200 with step 10
mean_rmse_per_depth = {}  # To store the mean RMSE for each max_depth

# Loop through each max_depth value
for max_depth in max_depth_values:
    rmse_values = []  # To store the RMSE for each n_estimators
    for n_estimators in n_estimators_values:
        rf_model = RandomForestRegressor(n_estimators=n_estimators,
                                         max_depth=max_depth,
                                         random_state=1,
                                         n_jobs=-1)
        rf_model.fit(X_train, y_train)

        # Predict on the validation set
        y_pred = rf_model.predict(X_val)

        # Calculate RMSE
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        rmse_values.append(rmse)

    # Calculate the mean RMSE for this max_depth
    mean_rmse = np.mean(rmse_values)
    mean_rmse_per_depth[max_depth] = mean_rmse
    print(f"max_depth: {max_depth}, Mean RMSE: {mean_rmse}")

# Find the best max_depth (lowest mean RMSE)
best_max_depth = min(mean_rmse_per_depth, key=mean_rmse_per_depth.get)
print(f"The best max_depth is: {best_max_depth} with a mean RMSE of {mean_rmse_per_depth[best_max_depth]}")
```

We can extract feature importance information from tree-based models.

At each step of the decision tree learning algorithm, it finds the best split. When doing it, we can calculate "gain" - the reduction in impurity before and after the split. This gain is quite useful in understanding what are the important features for tree-based models.

In Scikit-Learn, tree-based models contain this information in the feature_importances_ field.

For this homework question, we'll find the most important feature:

Train the model with these parameters:
n_estimators=10,
max_depth=20,
random_state=1,
n_jobs=-1 (optional)
Get the feature importance information from this model
What's the most important feature (among these 4)?

study_hours_per_week
attendance_rate
distance_to_school
teacher_quality

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the model with the given parameters
model = RandomForestRegressor(n_estimators=10,
                              max_depth=20,
                              random_state=1,
                              n_jobs=-1)

# Train the model
model.fit(X_train, y_train)

# Get the feature importance values
feature_importances = model.feature_importances_

# Use vec.feature_names_ to get the feature names from DictVectorizer
features = vec.feature_names_

# Create a dictionary of feature names and their corresponding importance
feature_importance_dict = dict(zip(features, feature_importances))

# Print the feature importance
print("Feature importances:", feature_importance_dict)

# Find the most important feature
most_important_feature = max(feature_importance_dict, key=feature_importance_dict.get)
print("Most important feature:", most_important_feature)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

# Convert your data into DMatrix format for XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)

# Create a watchlist to evaluate the model on the validation set during training
watchlist = [(dtrain, 'train'), (dval, 'validation')]
```

```python
# Define parameters with eta = 0.3
xgb_params_0_3 = {
    'eta': 0.3,
    'max_depth': 6,
    'min_child_weight': 1,
    'objective': 'reg:squarederror',
    'nthread': 8,
    'seed': 1,
    'verbosity': 1,
}

# Train the model
model_0_3 = xgb.train(xgb_params_0_3, dtrain, num_boost_round=100, evals=watchlist, early_stopping_rounds=10)
```

```python
# Define parameters with eta = 0.1
xgb_params_0_1 = {
    'eta': 0.1,
    'max_depth': 6,
    'min_child_weight': 1,
    'objective': 'reg:squarederror',
    'nthread': 8,
    'seed': 1,
    'verbosity': 1,
}

# Train the model
model_0_1 = xgb.train(xgb_params_0_1, dtrain, num_boost_round=100, evals=watchlist, early_stopping_rounds=10)
```

```python
# Get the best RMSE scores from each model
rmse_0_3 = model_0_3.best_score
rmse_0_1 = model_0_1.best_score

print(f"Best RMSE with eta=0.3: {rmse_0_3}")
print(f"Best RMSE with eta=0.1: {rmse_0_1}")
```
